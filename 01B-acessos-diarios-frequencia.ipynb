```python
import pandas as pd
import numpy as np
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date
```

```python
df_eventos = pd.read_csv('Data/01-eventos-consolidados_apenas-free.csv', index_col=0)
df_eventos.head()
```

```python
df_perfils = pd.read_csv('Data/02-perfils-higienizados.csv', index_col=0)
```

### Vamos fazer um merge nas tabelas

```python
df = df_perfils.merge(df_eventos,on='distinct_id')
df.head()
```

### Agora criar duas variáveis para separar hora de data de acesso

```python
df['Dates'] = pd.to_datetime(df['time']).dt.date
df['Time'] = pd.to_datetime(df['time']).dt.time
```

```python
df.head()
```

### Agrupar por ID e contar os valores de data únicos

```python
df_frequency = df.groupby('distinct_id').Dates.nunique().reset_index()
```
